```python
import pandas as pd

df = pd.read_csv("../../Dataset.csv")
print(df.head())
```

```python
# Length of restaurant name
df['Name_length'] = df['Restaurant Name'].apply(len)

# Length of address
df['Address_length'] = df['Address'].apply(len)

# Number of cuisines offered (split by comma)
df['Cuisine_count'] = df['Cuisines'].apply(lambda x: len(str(x).split(',')))
```

```python
# Encode Has Table booking (Yes/No -> 1/0)
df['Has_Table_Booking_encoded'] = df['Has Table booking'].apply(lambda x: 1 if x == 'Yes' else 0)

# Encode Has Online delivery (Yes/No -> 1/0)
df['Has_Online_Delivery_encoded'] = df['Has Online delivery'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
print(df[['Restaurant Name', 'Name_length', 
          'Address', 'Address_length', 
          'Cuisines', 'Cuisine_count', 
          'Has Table booking', 'Has_Table_Booking_encoded', 
          'Has Online delivery', 'Has_Online_Delivery_encoded']].head())
```

```python
import os
current_dir = os.getcwd()
output_path = os.path.join(current_dir, "Dataset_Feature_Engineered.csv")
df.to_csv(output_path, index=False)

print("Saved at:", output_path)
```

```python
import matplotlib.pyplot as plt

# 1 Histogram of restaurant name lengths
plt.hist(df['Name_length'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Restaurant Name Lengths")
plt.xlabel("Name Length")
plt.ylabel("Frequency")
plt.show()

# 2 Histogram of address lengths
plt.hist(df['Address_length'], bins=30, color='lightgreen', edgecolor='black')
plt.title("Distribution of Address Lengths")
plt.xlabel("Address Length")
plt.ylabel("Frequency")
plt.show()

# 3 Cuisine count distribution
plt.hist(df['Cuisine_count'], bins=20, color='salmon', edgecolor='black')
plt.title("Distribution of Cuisine Count")
plt.xlabel("Number of Cuisines")
plt.ylabel("Frequency")
plt.show()

# 4 Bar chart for Table Booking availability
df['Has_Table_Booking_encoded'].value_counts().plot(kind='bar', color=['orange','blue'])
plt.title("Has Table Booking (0=No, 1=Yes)")
plt.xlabel("Table Booking Availability")
plt.ylabel("Count")
plt.show()

# 5 Bar chart for Online Delivery availability
df['Has_Online_Delivery_encoded'].value_counts().plot(kind='bar', color=['purple','cyan'])
plt.title("Has Online Delivery (0=No, 1=Yes)")
plt.xlabel("Online Delivery Availability")
plt.ylabel("Count")
plt.show()
```
